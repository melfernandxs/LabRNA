```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

from sklearn.metrics import mean_squared_error
```

## 1

```python
print('Carregando Arquivo de teste')
arquivo = np.load('teste1.npy')
x = arquivo[0]
y = np.ravel(arquivo[1])

regr = MLPRegressor(hidden_layer_sizes=(2),
                    max_iter=100,
                    activation='relu', #{'identity', 'logistic', 'tanh', 'relu'},
                    solver='adam',
                    learning_rate = 'adaptive',
                    n_iter_no_change=50)
print('Treinando RNA')
regr = regr.fit(x,y)

print('Preditor')
y_est = regr.predict(x)

plt.figure(figsize=[14,7])

#plot curso original
plt.subplot(1,3,1)
plt.plot(x,y)

#plot aprendizagem
plt.subplot(1,3,2)
plt.plot(regr.loss_curve_)

#plot regressor
plt.subplot(1,3,3)
plt.plot(x,y,linewidth=1,color='yellow')
plt.plot(x,y_est,linewidth=2)
```

# Calculo de mdia e desvio padro

## 2

```python
print('Carregando Arquivo de teste')
arquivo = np.load('teste2.npy')
x = arquivo[0]
y = np.ravel(arquivo[1])

arquiteturas = [(2,), (5, 3), (10, 5)]

for arquitetura in arquiteturas:
    erros = []
    for i in range(10):
        print(f'Simulao numero {i+1} para a arquitetura {arquitetura}')
        
        regr = MLPRegressor(hidden_layer_sizes=arquitetura,
                            max_iter=1000,
                            activation='relu',
                            solver='adam',
                            learning_rate='adaptive',
                            n_iter_no_change=50)
        
        print('Treinando RNA')
        regr = regr.fit(x, y)
        
        print('Preditor')
        y_est = regr.predict(x)
        
        erro = mean_squared_error(y, y_est)
        erros.append(erro)
    
    # calculo da media e do desvio padro
    media_erro = np.mean(erros)
    desvio_padrao_erro = np.std(erros)
    
    print(f"Arquitetura {arquitetura}: Media = {media_erro}, Desvio padro = {desvio_padrao_erro}")

    plt.figure(figsize=[14,7])

    plt.subplot(1, 3, 1)
    plt.plot(x, y)

    plt.subplot(1, 3, 2)
    plt.plot(regr.loss_curve_)

    plt.subplot(1, 3, 3)
    plt.plot(x, y, linewidth=1, color='yellow')
    plt.plot(x, y_est, linewidth=2)

    plt.show()
```
